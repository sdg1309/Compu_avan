```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# --- 6. EJERCICIO PARA ESTUDIANTES ---
print("\n" + "EJERCICIO PRÁCTICO".center(60, "~"))
print("A continuación se muestran datos de tiempo de atención (minutos) a clientes:")
tiempos = [8.2, 5.7, 10.3, 7.8, 9.1, 12.5, 6.4, 11.2, 8.9, 14.1, 
           9.5, 13.2, 7.1, 10.8, 8.5, 6.9, 12.8, 11.7, 9.9, 15.3]

df = pd.DataFrame(tiempos, columns=['tiempos'])

print(tiempos)
print("\nInstrucciones:")
print("1. Calcula media, mediana y moda")

media = np.mean(tiempos)
mediana = np.median(tiempos)
moda = stats.mode(tiempos, keepdims=True)[0][0]

print("="*60)
print(f"- Media (Promedio): {media:.2f}")
print(f"- Mediana: {mediana:.2f}")
print(f"- Moda: {moda:.2f}")
print("="*60)

print("2. Calcula rango, varianza, desviación estándar e IQR")

rango = np.ptp(tiempos)  # Rango total
varianza = np.var(tiempos, ddof=1)
desviacion = np.std(tiempos, ddof=1)
q1, q3 = np.percentile(tiempos, [25, 75])
iqr = q3 - q1

print("="*60)
print(f"- Rango: {rango:.2f}")
print(f"- Varianza: {varianza:.2f}")
print(f"- Desviación Estándar: {desviacion:.2f}")
print(f"- Rango Intercuartílico (IQR): {iqr:.2f}")
print("="*60)

print("3. Crea un histograma y un boxplot")

# Histograma
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

sns.histplot(df['tiempos'], bins=15, kde=True, color='skyblue', ax=ax1)
ax1.axvline(media, color='r', linestyle='--', label=f'Media ({media:.2f})')
ax1.axvline(mediana, color='g', linestyle='-', label=f'Mediana ({mediana:.2f})')
ax1.axvline(moda, color='b', linestyle='-.', label=f'Moda ({moda:.2f})')
ax1.set_title('Distribución de tiempos - Histograma', fontsize=14)
ax1.set_xlabel('Tiempo')
ax1.set_ylabel('Frecuencia')
ax1.legend()

# Boxplot
sns.boxplot(x=df['tiempos'], color='lightgreen', ax=ax2)
ax2.set_title('Distribución de tiempos - Boxplot', fontsize=14)
ax2.set_xlabel('Tiempos')

plt.tight_layout()
plt.show()
```

```python
print("4. Interpreta los resultados: ¿Cómo es la distribución? ¿Hay outliers?")

print("Viendo el histograma, podemos inferir que la distribucion tiene un outlier, en el intervalo antes del tiempo 14.")

print("5. ¿Qué medida de tendencia central es más representativa y por qué?")

print("La tendendencia central mas representativa es la mediana, ya que es mas conservativa, se acerta mas al pico mas alto de la grafica.")
```
