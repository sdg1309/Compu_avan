```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

# Crear datos:

```python
np.random.seed(43)
x1 = np.random.normal(100,10,100) # media, desviacion estandar, cantidad de datos
x2 = np.random.normal(0,100,100)
data = np.array([x1,x2]).T
```

```python
plt.figure(figsize=(8,6))
plt.title("Data original",size=15)
plt.boxplot(data)
plt.xticks([1, 2], ['x1','x2'],size=12)
plt.show()
```

# Estandarizar datos:

```python
data_standard = StandardScaler().fit_transform(data) # se calcula mean y std, luego se aplica la formula
print(f"mean: {data_standard.mean(axis=0).round(1)}")
print(f"std: {data_standard.std(axis=0)}")
```

```python
plt.figure(figsize=(8,6))
plt.title("Data estandarizada", size=15)
plt.boxplot(data_standard)
plt.xticks([1, 2], ['x1','x2'], size=12)
plt.show()
```

# Normalizar datos:

```python
data_norm = MinMaxScaler().fit_transform(data) # se calcula maximo y minimo, luego se aplica la formula
print(f"min: {data_norm.min(axis=0)}")
print(f"max: {data_norm.max(axis=0)}")
```

```python
plt.figure(figsize=(8,6))
plt.title("Data normalizada", size=15)
plt.boxplot(data_norm)
plt.xticks([1, 2], ['x1','x2'], size=12)
plt.show()
```
