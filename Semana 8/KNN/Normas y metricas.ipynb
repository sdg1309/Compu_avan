```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.random.normal([2,5],[1,1.5],size=(5000,2))
X
```

```python
plt.scatter(*X.T,alpha=0.3,color='darkblue')
plt.show()
```

```python
centro = np.mean(X,axis=0)
centro
```

```python
d1 = np.linalg.norm(X-centro,ord=1,axis=1)
d2 = np.linalg.norm(X-centro,ord=2,axis=1)
dInf = np.linalg.norm(X-centro,ord=np.inf,axis=1)
d1.shape
```

```python
def visualizar(distancias,pertencil):
    corte = np.percentile(distancias,pertencil)
    filtro = (distancias<corte)
    plt.scatter(*X.T,alpha=0.5,color='darkblue')
    plt.scatter(*X[filtro].T,alpha=0.2,color='red')
    plt.show()
```

```python
visualizar(d1,80)
```

```python
visualizar(d2,80)
```

```python
visualizar(dInf,80)
```
