```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
data = load_breast_cancer()
data.keys()
```

```python
data['data'].shape
```

```python
data['feature_names']
```

```python
pd.DataFrame(data['data'],columns=data['feature_names']).describe().T
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x=data.target)
plt.show()
```

```python
X_standardized = StandardScaler().fit_transform(data['data'])
X_train, X_test, y_train, y_test = train_test_split(
    X_standardized, data['target'], test_size=0.3, random_state=17)
```

```python
list_scores = []
range_k = range(1,21)
for k in range_k:
    knn = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    train_score = knn.score(X_train, y_train)
    test_score = knn.score(X_test,y_test)
    list_scores.append((train_score,test_score))
```

```python
df_scores = pd.DataFrame(list_scores,columns = ['Train','Test'],index=range_k)
df_scores.index.name = 'k'
df_scores
```

```python
df_scores.plot.line(y=['Train','Test'],figsize=(10,6),fontsize=13)
plt.legend(fontsize=14)
plt.xlabel('K',size=13)
plt.ylabel('Accuracy',size=13)
plt.xticks(range_k)
plt.grid()
plt.show()
```
