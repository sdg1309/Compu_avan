```python
# --- 1. IMPORTAR LIBRERAS ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configuracin de estilos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
np.random.seed(42)  # Para reproducibilidad

# --- 2. DATOS DE EJEMPLO ---
# Crear datos simulados: salarios mensuales (en $) de 100 empleados
salarios = np.concatenate([
    np.random.normal(1200, 100, 70),    # Mayora de empleados
    np.random.normal(2500, 150, 20),    # Gerentes
    np.random.normal(5000, 500, 10)     # Directivos
])

# Crear un DataFrame
df = pd.DataFrame(salarios, columns=['Salario'])

# --- 3. CLCULO DE MEDIDAS ESTADSTICAS ---
# Tendencia central
media = np.mean(salarios)
mediana = np.median(salarios)
moda = stats.mode(salarios, keepdims=True)[0][0]

# Dispersin
rango = np.ptp(salarios)  # Rango total
varianza = np.var(salarios, ddof=1)
desviacion = np.std(salarios, ddof=1)
q1, q3 = np.percentile(salarios, [25, 75])
iqr = q3 - q1

# --- 4. VISUALIZACIONES ---
# Configurar figura con 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Histograma
sns.histplot(df['Salario'], bins=15, kde=True, color='skyblue', ax=ax1)
ax1.axvline(media, color='r', linestyle='--', label=f'Media (${media:.2f})')
ax1.axvline(mediana, color='g', linestyle='-', label=f'Mediana (${mediana:.2f})')
ax1.axvline(moda, color='b', linestyle='-.', label=f'Moda (${moda:.2f})')
ax1.set_title('Distribucin de Salarios - Histograma', fontsize=14)
ax1.set_xlabel('Salario Mensual ($)')
ax1.set_ylabel('Frecuencia')
ax1.legend()

# Boxplot
sns.boxplot(x=df['Salario'], color='lightgreen', ax=ax2)
ax2.set_title('Distribucin de Salarios - Boxplot', fontsize=14)
ax2.set_xlabel('Salario Mensual ($)')

# Aadir anotaciones con medidas de dispersin
plt.figtext(0.5, 0.01, 
            f"Rango: ${rango:.2f} | Varianza: ${varianza:.2f} | Desv. Estndar: ${desviacion:.2f} | IQR: ${iqr:.2f}",
            ha="center", fontsize=12, bbox={"facecolor":"orange", "alpha":0.2, "pad":5})

plt.tight_layout()
plt.show()

# --- 5. PRESENTAR RESULTADOS NUMRICOS ---
print("\n" + "="*60)
print("RESUMEN ESTADSTICO DE LOS SALARIOS".center(60))
print("="*60)
print(f"- Media (Promedio): ${media:.2f}")
print(f"- Mediana: ${mediana:.2f}")
print(f"- Moda: ${moda:.2f}")
print(f"- Rango: ${rango:.2f}")
print(f"- Varianza: ${varianza:.2f}")
print(f"- Desviacin Estndar: ${desviacion:.2f}")
print(f"- Rango Intercuartlico (IQR): ${iqr:.2f}")
print("="*60)
```
