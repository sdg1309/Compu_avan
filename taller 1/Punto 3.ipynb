```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Cargar dataset
wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)
y = wine.target

# Descripcin estadstica
print(X.describe())

# Histogramas
X.hist(figsize=(15, 10), bins=20)
plt.suptitle("Histogramas de caractersticas del vino", fontsize=16)
plt.show()
```

```python
# Matriz de correlacin
plt.figure(figsize=(12, 8))
sns.heatmap(X.corr(), annot=False, cmap="coolwarm")
plt.title("Matriz de correlacin - Wine")
plt.show()

# Scatter plot de dos variables relevantes
sns.scatterplot(x="alcohol", y="color_intensity", hue=y, palette="viridis", data=X)
plt.title("Alcohol vs Intensidad de color")
plt.show()
```

```python
# Escalar antes de PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

print("Varianza explicada por cada componente:", pca.explained_variance_ratio_)
print("Varianza acumulada:", np.cumsum(pca.explained_variance_ratio_))

# 2D
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap="viridis")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA 2D - Wine dataset")
plt.show()

# 3D
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X_pca[:,0], X_pca[:,1], X_pca[:,2], c=y, cmap="viridis")
ax.set_xlabel("PC1"); ax.set_ylabel("PC2"); ax.set_zlabel("PC3")
ax.set_title("PCA 3D - Wine dataset")
plt.show()
```

```python
inertias = []
silhouettes = []
K = range(2, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouettes.append(silhouette_score(X_scaled, labels))

# Mtodo del codo
plt.plot(K, inertias, "-o")
plt.xlabel("Nmero de clusters (k)")
plt.ylabel("Inercia")
plt.title("Mtodo del Codo - Wine")
plt.show()

# Silueta
plt.plot(K, silhouettes, "-o")
plt.xlabel("Nmero de clusters (k)")
plt.ylabel("Score de Silueta")
plt.title("Mtodo de Silueta - Wine")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

plt.scatter(X_pca[:,0], X_pca[:,1], c=clusters, cmap="plasma")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clustering K-Means en PCA 2D (Wine)")
plt.show()
```
